# Iris dataset

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
names =  ['sepal_length','sepal_width','petal_length','petal_width','class']
```

```python
iris = pd.read_csv('iris.data', names=names)
```

```python
iris.head()
```

```python
iris['class'].value_counts()
```

# 2D Scatter plot

```python
ax = sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='class')
plt.show()
```
