```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.FacetGrid(data=tips, row='smoker', col='day', col_order=['Sun', 'Sat','Fri', 'Thur'])\
        .map(sns.distplot, 'total_bill')
plt.show()
```
