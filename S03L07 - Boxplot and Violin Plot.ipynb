```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
names =  ['sepal_length','sepal_width','petal_length','petal_width','class']
iris = pd.read_csv('iris.data', names=names)
iris.head()
```

```python

sns.boxplot(data=iris, x='petal_length', y='class')
plt.show()
```

```python
sns.violinplot(data=iris, x='petal_length', y='class')
fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.show()
```
