# Uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(0,10,100000)
plt.hist(values,50)
plt.show()
```

# Normal distribution

```python
values = np.random.normal(0,1.5,100000)
plt.hist(values,50)
plt.show()
```

# Exponenetial distribution

```python
from scipy.stats import expon
x = range(0,10)
#print(expon.pdf(x))
plt.plot(x,expon.pdf(x))
plt.show()
```

# Probability density funtion

```python
import pandas as pd
import seaborn as sb
sb.set()

from matplotlib import pyplot as plt
values = np.random.normal(0,1,200)
sb.distplot(values)
plt.show()
```

# Probability mass function

```python
x1 = np.array([1,2,3,4,5])
x_name = ['A brand', 'B brand', 'C brand', 'D brand', 'E brand']
y1 = [55,85,96,88,99]
plt.bar(x1,y1,color = 'blue')

x_pmf = np.insert(x1,[0,5],[0,6])
y_pmf = np.insert(y1,[0,5],[0,0])
 

plt.plot(x_pmf,y_pmf)
plt.title('Demo graphs')
plt.xlabel('X-axis')
plt.xticks(x1, x_name)
plt.ylabel('Y-axis')
plt.show()
```
