```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
names =  ['sepal_length','sepal_width','petal_length','petal_width','class']
iris = pd.read_csv('iris.data', names=names)
iris.head()
```

```python
sns.FacetGrid(data=iris,hue='class', height=5).map(sns.distplot, 'sepal_length').add_legend()
plt.show()
```

```python
sns.FacetGrid(data=iris,hue='class', height=5).map(sns.distplot, 'sepal_width').add_legend()
plt.show()
```

```python
sns.FacetGrid(data=iris,hue='class', height=5).map(sns.distplot, 'petal_length').add_legend()
plt.show()
```

```python
sns.FacetGrid(data=iris,hue='class', height=5).map(sns.distplot, 'petal_width').add_legend()
plt.show()
```
