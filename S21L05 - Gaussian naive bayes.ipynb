```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
plt.rcParams["figure.figsize"] = (15,8)


mu = 40
variance =30
sigma = math.sqrt(variance)
women = np.linspace(mu - 3*sigma, mu + 5*sigma, 100)
plt.plot(women, stats.norm.pdf(women, mu, sigma),linewidth=7.0,color="green")


mmu = 55
mvariance = 35
msigma = math.sqrt(mvariance)
men = np.linspace(mmu - 5*msigma, mmu + 2*msigma, 100)
plt.plot(men, stats.norm.pdf(men, mmu, msigma),linewidth=7.0,color="red")


plt.show()
```
